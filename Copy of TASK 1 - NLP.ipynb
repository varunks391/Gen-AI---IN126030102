Task 1: Conceptual Understanding

1.Difference between "Love" and "love" in NLP
--> “Love” (capitalized) might appear at the start of a sentence or as a proper noun (e.g., a brand name, title).
--> “love” (lowercase) usually refers to the emotion or verb.


2.What happens if stopwords are not removed?
- Stopwords are common words like the, is, at, and.
- If not removed:
- They increase dimensionality of the dataset unnecessarily.
- They add noise, making models less efficient.
- They can slow down training and reduce accuracy in tasks like topic modeling or keyword extraction.





3.Real-world scenarios where removing stopwords can be harmful
- Sentiment Analysis
- Words like not, never, no are often considered stopwords but are critical for polarity.
- Example: “I do not like this movie” → removing not changes the meaning entirely.
- Question Answering / Search Engines
- Stopwords like who, what, where, when are essential for understanding the type of query.
- Example: “Where is the Eiffel Tower?” → removing where makes the query incomplete.

4.Difference between Stemming and Lemmatization
- Stemming: Mechanical chopping of word endings.
- Lemmatization: Considers context and part of speech to return the correct base form (lemma).


Task 2:Build Advanced Preprocessing Function

```python
import re
def preprocess_text(text):
    # 1. Convert to lowercase
    text = text.lower()
    # 2. Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)
    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)
    # 4. Handle repeated characters (reduce to max 2)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    # 5. Remove extra spaces and punctuation clutter
    text = re.sub(r'\s+', ' ', text).strip()
    # 6. Tokenize
    tokens = text.split()
    # 7. Remove very short tokens (≤ 2), but keep "no" and "not"
    meaningful = {"no", "not"}
    tokens = [t for t in tokens if len(t) > 2 or t in meaningful]
    # 8. Rejoin tokens
    return " ".join(tokens)
    print(preprocess_text("I have 2 dogs"))
# Output: "i have dogs"

print(preprocess_text("soooo goooood!!!"))
# Output: "so good"

print(preprocess_text("WOW!!! This is GREAT!!!"))
# Output: "wow this is great"

print(preprocess_text("Visit http://example.com now"))
# Output: "visit now"

print(preprocess_text("no way!!! not possible"))
# Output: "no way not possible"
```

Task 4: Token Analytics


```python
import re
import numpy as np

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)   # remove URLs/emails
    text = re.sub(r'\d+', '', text)                                # remove numbers
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)                     # reduce repeated chars
    text = re.sub(r'\s+', ' ', text).strip()                       # remove extra spaces
    tokens = text.split()
    meaningful = {"no", "not"}
    tokens = [t for t in tokens if len(t) > 2 or t in meaningful]  # remove short tokens
    return tokens, " ".join(tokens)

# Sample inputs
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Token Analytics
print("=== Token Analytics ===\n")
analytics = []
for s in sample_inputs:
    tokens, cleaned = preprocess_text(s)
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = np.mean([len(t) for t in tokens]) if tokens else 0
    analytics.append((cleaned, total, unique, avg_len))
    print(f"Original: {s}")
    print(f"Cleaned: {cleaned}")
    print(f"Total Tokens: {total}, Unique Tokens: {unique}, Avg Token Length: {avg_len:.2f}\n")

# Analysis Questions
most_noise = min(analytics, key=lambda x: x[1])  # fewest tokens retained
most_meaningful = max(analytics, key=lambda x: x[2])  # most unique tokens retained

print("=== Analysis ===")
print(f"Sentence with most noise removed: '{most_noise[0]}'")
print(f"Sentence retaining most meaningful tokens: '{most_meaningful[0]}'")
```

Task 5: Frequency Analysis

```python
from collections import Counter
import re
import numpy as np

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)   # remove URLs/emails
    text = re.sub(r'\d+', '', text)                                # remove numbers
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)                     # reduce repeated chars
    text = re.sub(r'\s+', ' ', text).strip()                       # remove extra spaces
    tokens = text.split()
    meaningful = {"no", "not"}
    tokens = [t for t in tokens if len(t) > 2 or t in meaningful]  # remove short tokens
    return tokens

# Sample inputs (same as Task 3)
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Collect all tokens
all_tokens = []
for s in sample_inputs:
    all_tokens.extend(preprocess_text(s))

# Frequency analysis
counter = Counter(all_tokens)

print("=== Top 10 Most Frequent Words ===")
for word, freq in counter.most_common(10):
    print(f"{word}: {freq}")

print("\n=== Top 5 Least Frequent Words ===")
for word, freq in counter.most_common()[-5:]:
    print(f"{word}: {freq}")
```

Task 6: Build Full Pipeline

```python
import re

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (reduce to max 2)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    tokens = text.split()

    # Remove very short tokens (≤ 2), but keep "no" and "not"
    meaningful = {"no", "not"}
    tokens = [t for t in tokens if len(t) > 2 or t in meaningful]

    return tokens, " ".join(tokens)


def full_pipeline(text_list):
    all_tokens = []
    cleaned_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        cleaned_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": cleaned_sentences
    }
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

result = full_pipeline(sample_inputs)
print(result)
```

```python
import re

def preprocess_text(text):
    # Handle None or empty string
    if not text or text.strip() == "":
        return [], ""

    # Convert to lowercase
    text = text.lower()

    # Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (reduce to max 2)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove emojis and non-alphanumeric symbols (optional safety)
    text = re.sub(r'[^\w\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    tokens = text.split()

    # Remove very short tokens (≤ 2), but keep "no" and "not"
    meaningful = {"no", "not"}
    tokens = [t for t in tokens if len(t) > 2 or t in meaningful]

    # If no tokens remain, return empty safely
    if not tokens:
        return [], ""

    return tokens, " ".join(tokens)
print(preprocess_text(""))
# Output: ([], "")

print(preprocess_text("😍😍😍😍"))
# Output: ([], "")

print(preprocess_text("1234567890"))
# Output: ([], "")
```
